```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
plt.figure()
plt.plot (df["Spending Score (1-100)"])
plt.xlabel("Customer Index")
plt.ylabel("Spending Score")
plt.title("Line plot Spending Score")
plt.show()
df
```

```python
plt.figure()
plt.scatter(df["Age"],df["Spending Score (1-100)"])
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Age vs Spending Score")
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()
plt.figure()
plt.bar(gender_counts.index,gender_counts.values)
plt.xlabel("Gender")
plt.ylabel("Numer of Customers")
plt.title("Gender Distribution")
plt.show()
        
```

```python
gender_counts = df["Gender"].value_counts()

plt.figure()
plt.pie(gender_counts.values, labels=gender_counts.index, 
        autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure()
plt.hist(df["Age"],bins=5)
plt.xlabel("Age")
plt.ylabel("count")
plt.title("Age Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].hist(df["Age"], bins=10)
axs[0, 0].set_title("Age Distribution")


axs[0, 1].hist(df["Annual Income (k$)"], bins=10)
axs[0, 1].set_title("Annual Income Distribution")


axs[1, 0].hist(df["Spending Score (1-100)"], bins=10)
axs[1, 0].set_title("Spending Score Distribution")


gender_counts = df["Gender"].value_counts()
axs[1, 1].bar(gender_counts.index, gender_counts.values)
axs[1, 1].set_title("Gender Count")

plt.tight_layout()
plt.show()
```
