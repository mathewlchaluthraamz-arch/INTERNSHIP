```python
import seaborn as sns
df = sns.load_dataset("iris")
sns.scatterplot(data=df,x='sepal_width', y='petal_length', hue='species')
plt.title("Scatter plot:sepal width vs petal length")
plt.show()
```

```python
sns.lineplot(data=df,x='species', y='petal_length', )
plt.title("Line plot:Species vs petal length")
plt.show()
```

```python
sns.barplot(x='species', y='petal_length', data=df )
plt.title("Average Length by species")
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df )
plt.title("Average Length by species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
sns.kdeplot(df['petal_length'],shade=True,color='yellow')
plt.title('KDE Plot of Petal Length by Species')
plt.xlabel('petal_length')
plt.ylabel('Density')
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=df )
plt.title("Violin Plot of petal Length by Species")
plt.show()
```

```python
sns.histplot(df['petal_length'], kde=True )
plt.title("Distribution Plot:Petal length")
plt.show()
```

```python
sns.pairplot(df, hue='species' )
plt.title("pairwise Plot:Iris Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```
