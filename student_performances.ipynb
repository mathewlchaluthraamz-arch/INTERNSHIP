```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("student_performance.csv")
df.head()
```

```python
plt.figure()
plt.plot(df["student_id"], df["total_score"])

plt.xlabel("Student ID")
plt.ylabel("Total Score")
plt.title("Student Total Score Line Plot")

plt.show()
df
```

```python
plt.figure()
plt.scatter(df["weekly_self_study_hours"], df["total_score"])

plt.xlabel("Weekly self study hours")
plt.ylabel("Total Score")
plt.title("Attendance vs Total Score")

plt.show()
```

```python

grade_counts = df["grade"].value_counts()

plt.figure()
plt.bar(grade_counts.index, grade_counts.values)

plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.title("Grade Distribution")

plt.show()
```

```python

grade_counts = df["grade"].value_counts()

plt.figure()
plt.pie(grade_counts.values,
        labels=grade_counts.index,
        autopct='%1.1f%%')

plt.title("Grade Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0,0].hist(df["weekly_self_study_hours"],bins=10)
axs[0,0].set_title("Students study hours")

axs[0,1].hist(df["attendance_percentage"],bins=10)
axs[0,1].set_title("Attendences percentage")

axs[1,0].hist(df["class_participation"],bins=10)
axs[1,0].set_title("Class participation")

axs[1,1].hist(df["total_score"],bins=10)
axs[1,1].set_title("Total score")

plt.tight_layout()
plt.show()
```
