```python
sns.scatterplot(data=df,x='total_bill', y='tip', hue='day')
plt.title("Scatter plot:Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=df,x='total_bill', y='tip', )
plt.title("Line plot:Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df )
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df )
plt.title("Average Total Bill by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
sns.kdeplot(df['total_bill'],shade=True,color='yellow')
plt.title('Kernel Density Estimate (KDE) of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df )
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.histplot(df['total_bill'], kde=True )
plt.title("Distribution Plot:Total Bill")
plt.show()
```

```python
sns.pairplot(df, hue='sex' )
plt.title("pairwise Plot:Tips Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```
